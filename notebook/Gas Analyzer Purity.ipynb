```python
%matplotlib notebook
from IPython import display
import seaborn as sns
from runco import config, sc
import datetime
import matplotlib.pyplot as plt
import numpy
from IPython import display
import time
sns.set_context('talk')
sns.set_style('whitegrid')
```

```python
gas_analyzers={
    'n2':{'var':'uB_Cryo_IFIX_1_0/AT610',
         'unit':'ppm'},
    'o2':{'var':'uB_Cryo_IFIX_1_0/AT609',
         'unit':'ppb'},
    'h2o':{'var':'uB_Cryo_IFIX_1_0/AT608',
         'unit':'ppb'},
}
```

```python
conf = config.Config()
slowcon = sc.ConfiguredSlowControls(conf.data['slowcontrols'])
form = lambda x : x.strftime("%Y-%m-%d %H:%M:%S")
```

```python
for key in gas_analyzers:
    ch = slowcon.channel_id_by_name[gas_analyzers[key]['var']]
    gas_analyzers[key]['ch'] = ch
```

```python
now = datetime.datetime.now()+datetime.timedelta(hours=5)
one_week_ago = now - datetime.timedelta(days=6)
key='n2'
ax = plt.figure()
slowcon.query_timebinned_data(gas_analyzers[key]['ch'],1, form(one_week_ago), form(now))
result = slowcon.cur.fetchall()
x = [datetime.datetime.fromtimestamp(i[1])+datetime.timedelta(hours=5) for i in result]
y = [i[2] for i in result]
plt.plot(x,y)
plt.xlabel('Date/Time')
plt.ylabel('{} Purity [{}]'.format(key, gas_analyzers[key]['unit']))
ax.autofmt_xdate()
plt.show()
```

```python
key='o2'
ax = plt.figure()
slowcon.query_timebinned_data(gas_analyzers[key]['ch'],1, form(one_week_ago), form(now))
result = slowcon.cur.fetchall()
x = [datetime.datetime.fromtimestamp(i[1])+datetime.timedelta(hours=5) for i in result]
y = [i[2] for i in result]
plt.plot(x,y)
plt.xlabel('Date/Time')
plt.ylabel('{} Purity [{}]'.format(key, gas_analyzers[key]['unit']))
ax.autofmt_xdate()
plt.show()
```
